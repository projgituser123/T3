```python
import requests
import os
from bs4 import BeautifulSoup

url = "https://subslikescript.com/movies"
res = requests.get(url)
print("Staus code = ", res.status_code)

res=res.text

soup = BeautifulSoup(res, "html.parser")
print("HTML code = \n\n", soup.prettify())
```

```python
a = soup.find("ul", attrs={"scripts-list"})
link =[]
movie= []
movie_list = a.find_all("a", href = True) # this method returns a list datatype
print("List of all movie links = \n\n", movie_list)

```

```python
# now to extract the movie name and their links which are present in list datatype, we will use for loop to access those .... 
for i in movie_list :
    movie.append(i.text.strip())
    link.append("https://subslikescript.com/movie" + i["href"])

print("Movie names = \n", movie)
print("-"*50)
print("Links = \n", link)


# now create a df from these lists :::
import pandas as pd
df = pd.DataFrame({"Movie_name" : movie,
                   "Movie_link"  : link})
df
```
