```python
import requests
from bs4 import BeautifulSoup

url = "https://subslikescript.com/movie/Titanic-46435"
res = requests.get(url)
print("Staus code = ", res.status_code) # 200 = means "ok"
print("-" *50)
res = res.text # convert it into text file..
soup = BeautifulSoup(res, "html.parser") # make the object of the beautiful soup... "html.parser" = means convert the text response into hmtl code...
#print("HTML code = \n", soup)
print("HTML code = \n", soup.prettify()) # this method formats the html file properly...
```

```python
# now find the specific data : say for eg : titanic movie heading ...
print(soup.find("title"))
print("-" *50)
soup.find("title").text.strip() # in the form of text
```

```python
# find all the div tag :: 

#soup.find_all("div").text.strip() // you cannot apply text.strip() here because this is multiple line answer .. 
soup.find_all("div")
```

```python
# find the div tag and the specific class : 

print(soup.find("div", attrs={"subtitle-cue"}).text.strip())
print(soup.find("div", class_ = "subtitle-cue").text.strip())
# both method are same
```

```python
# now write the data into a newly created txt file : 
with open("D:/dvc/py/web scraping/txt files/titanic.txt", "w", encoding="utf-8") as file : 
    file.write(soup.find("div", attrs={"full-script"}).text.strip())
```
