```python
import os 
import requests
from bs4 import BeautifulSoup

url = "https://indianexpress.com/latest-news/"
res = requests.get(url=url)
print("Satus code = ", res.status_code)
res = res.text
soup = BeautifulSoup(res, "html.parser")
print("HTML code = \n\n", soup.prettify())
```

```python
nation_div = soup.find("div", class_ = "nation")
title_div = nation_div.find_all("div", class_= "title")

print(title_div)
print("-"*50)
news=[]
for i in title_div : 
    n = i.find("a", href = True).text.strip()
    news.append(n)

for i in news:
    print(i + "\n") 
```
