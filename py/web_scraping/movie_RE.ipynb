```python
import os
import requests
from bs4 import BeautifulSoup

url = "https://subslikescript.com/movie/Resident_Evil_Death_Island-26674627"
res = requests.get(url=url)
print("status code = ", res.status_code)
print("-"*50)
res = res.text
print("HTML code = \n\n", BeautifulSoup(res, "html.parser").prettify())
```

```python
cwd = os.getcwd() + "/txt_files/"
print(cwd)

# shorter method to set the path for txt ....
# if you dont want to store in "soup" variable then : BeautifulSoup(res, "html.parser") is your object and then you can use the its method instead of soup.find(--parameters--)
with open(cwd + "re.txt", "w+", encoding="utf-8") as file : 
    file.write(BeautifulSoup(res, "html.parser").find("div", attrs={"full-script"}).text.strip())
```
